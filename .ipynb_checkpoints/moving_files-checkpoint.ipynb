```python
import pydicom
import glob
import sys
files=[]
for fname in glob.glob("stage_2_train_mini/*", recursive=False):
    print("loading: {}".format(fname))
    try: 
        files.append(pydicom.read_file(fname))
    except: 
        print("Error with file "+fname)
```

```python
from collections import defaultdict
from shutil import copyfile
import os
# dict for counting number of files for each patient ID
#patient_id_count = defaultdict(lambda: 0)

for f in files:
    id = f.PatientID    # this gets the patient ID from the current file
    if os.path.exists("patient_sorted_train_mini/"+id):
        os.mkdir("patient_sorted_train_mini/"+id)
    copyfile(f.SOPInstanceUID, "patient_sorted_train_mini/"+id+"/"+f.file_name+".dcm")
    #patient_id_count{id} += 1
```
